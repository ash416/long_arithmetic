```python
def getData(nameL, nameB):
    g = open(nameL, 'r')
    dict1 = dict()
    dict2 = dict()
    for line in g:
        listline = line.split(' ')
        dict1.update({float(listline[1].split('\n')[0]) : float(listline[0])})
    g.close()
    g = open(nameB, 'r')
    for line in g:
        listline = line.split(' ')
        dict2.update({float(listline[1].split('\n')[0]) : float(listline[0])})
    g.close()

    dict1 = sorted(dict1.items())
    dict2 = sorted(dict2.items())
    x = []
    y1 = []
    y2 = []
    for i in range(len(dict1)):
        x.append(dict1[i][0])
        y1.append(dict1[i][1])
        y2.append(dict2[i][1])
    return x, y1, y2
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
def builtGraph(x, y1, y2, title, xlab, imgname):
    fig = plt.figure(figsize=(20, 10))
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel('время')

    plt.plot(x, y1, color = 'blue', linestyle = 'solid',
             label = 'little')
    plt.scatter(x, y1, color = 'blue')
    plt.plot(x, y2, color = 'red', linestyle = 'solid',
             label = 'big')
    plt.scatter(x, y2, color = 'red')

    plt.legend(loc = 'upper left', prop={'size': 26})
    plt.savefig(imgname)
   # plt.show()
```

```python
x, y1, y2 = getData('../results_1/resultsLsum.txt', '../results_1/resultsBsum.txt')
```

```python
builtGraph(x, y1, y2, 'Сумма двух чисел разной длины','размер суммы','../graphs_1/sum_time_lensum.png')
```

```python
builtGraph(x, y1, y2, 'Разность двух чисел разной длины','размер максимального числа','../graphs_1/diff_time_lenmaxnum.png')
```
