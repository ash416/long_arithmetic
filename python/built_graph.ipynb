```python
def getData(nameL, nameB):
    g = open(nameL, 'r')
    dict1 = dict()
    dict2 = dict()
    for line in g:
        listline = line.split(' ')
        dict1.update({float(listline[1].split('\n')[0]) : float(listline[0])})
    g.close()
    g = open(nameB, 'r')
    for line in g:
        listline = line.split(' ')
        dict2.update({float(listline[1].split('\n')[0]) : float(listline[0])})
    g.close()

    dict1 = sorted(dict1.items())
    dict2 = sorted(dict2.items())
    x = []
    y1 = []
    y2 = []
    for i in range(len(dict1)):
        x.append(dict1[i][0])
        y1.append(dict1[i][1])
        y2.append(dict2[i][1])
    return x, y1, y2
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.patches as mpatches
import seaborn as sns

def builtGraph(x, y1, y2, title, xlab, imgname):
    fig = plt.figure(figsize=(20, 10))
    # Добавление на рисунок прямоугольной (по умолчанию) области рисования
    plt.title(title, {'fontname':'Times New Roman'}, fontweight='bold', fontsize=40)
    plt.xlabel(xlab, {'fontname':'Times New Roman'}, fontweight='bold', fontsize=30)
    plt.ylabel('время, мс', {'fontname':'Times New Roman'}, fontweight='bold', fontsize=30)
    rc={'font.size': 20, 'axes.labelsize': 32, 'legend.fontsize': 32.0, 
    'axes.titlesize': 32, 'xtick.labelsize': 20, 'ytick.labelsize': 20}
    plt.rcParams.update(**rc)
    plt.grid(True)
    plt.plot(x, y1, color = 'blue', linestyle = 'solid',
             label = 'little')

    plt.plot(x, y2, color = 'red', linestyle = 'solid',
             label = 'big')

    plt.fill_between(x, y1, 0, color='blue', alpha=0.75) 
    plt.fill_between(x, y2, 0, color='red', alpha=0.25) 
    red_patch = mpatches.Patch(color='red', label='BIG ENDIAN')
    blue_patch = mpatches.Patch(color='blue', label='LITTLE ENDIAN')
    plt.legend(handles=[red_patch, blue_patch])
   # plt.legend(loc = 'upper left', prop={'size': 26})
    #plt.xlim(5000, 9000)
    plt.savefig(imgname)
    plt.show()
```

```python
x, y1, y2 = getData('../results_int_1/resultsLsum.txt', '../results_int_1/resultsBsum.txt')
```

```python
builtGraph(x, y1, y2, 'Сумма','длина суммы','../graphs_int_1/sum_time_lensum.png')
```

```python
builtGraph(x, y1, y2, 'Вычитание','длина максимального числа','../graphs_int_1/diff_time_maxnum.png')
```
