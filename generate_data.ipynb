```python
f = open('Div.txt', 'w')
j = 0
while j < 50:
    z = random.randint(1000, 10000)
    x = random.randint(100, 1000)
    num1 = 0
    k = 10
    if (z < x):
        continue
    else:
        j +=1
    for i in range(z):
        a = random.randint(0, 9)
        if i == 0 and a == 0:
            a += 1
        num1 = num1 * k + a

    num2 = 0
    k = 10
    for i in range(x):
        b = random.randint(0, 9)
        if i == 0 and b == 0:
            b += 1
        num2 = num2 * k + b
    res = int(num1 // num2) 
    f.write(str(num1) + ' ')
    f.write(str(num2) + ' ')
    f.write(str(res) + '\n')
j = 0
while j < 50:
    z = random.randint(100, 10000)
    x = random.randint(10, 1000)
    num1 = 0
    k = 10
    if (z < x):
        continue
    else:
        j +=1
    for i in range(z):
        a = random.randint(0, 9)
        if i == 0 and a == 0:
            a += 1
        num1 = num1 * k + a

    num2 = 0
    k = 10
    for i in range(x):
        b = random.randint(0, 9)
        if i == 0 and b == 0:
            b += 1
        num2 = num2 * k + b
    res = int(num1 // num2) 
    f.write(str(num1) + ' ')
    f.write(str(num2) + ' ')
    f.write(str(res) + '\n')

f.close()
```
